```python
import pandas as pd
```

```python
data = pd.read_csv('인하대역 이용승객 데이터.csv') 
#csv 파일의 경로를 지정해주거나, 이렇게 지정 후 이 ipynb파일과 동일 폴더에 저장해두면 됨

data.iloc[0:24]
```

```python
week = ['월', '화', '수', '목', '금']
day = input("무슨 요일인가요? : ")
while not day in week or day in data:
        day = input("요일을 제대로 입력해주세요 : ")
        if day in week or day in data:
            break
if day in week:
    day = day + '요일'

ending_time = int(input("수업 종료 시간(ex:1200) : "))
arrive_time = int(input("도착해야하는 시간(ex:1600) : "))
while ending_time > arrive_time:
    print("진심이면 외박밖에 답이 없습니다.")
    arrive_time = int(input("도착해야하는 시간을 제대로 입력해주세요 : "))
spending_time = int(input("소요되는 시간(ex:230 - 2시간 30분) : "))
```

```python
data['afterschool'] = data['이용시간'] >= ending_time #수업시간 이후
data['available'] = (arrive_time - spending_time) >= data['이용시간'] #최대 탑승 가능 시간 (도착시간-소요시간)
indexlist = []
for i in data.index: #i를 index만큼 반복
    if str(data['afterschool'][i]) == 'True': #수업시간 이후이면서
        if str(data['available'][i]) == 'True': #탑승이 가능한 열차
            indexlist.append(i) #i를 리스트에 추가
            
min_value = data[day][indexlist[0]:indexlist[-1]].min() #리스트 범위 내에서 최소값 서치

min_index = data.index[data[day]==min_value] #그 최소값의 인덱스 서치

print(min_value, min_index) #참고용. 최소값과 인덱스 프린트
```

```python
recommend = str(data['이용시간'].iloc[min_index].values) #추천 시간 (최소값에 대한 이용시간 데이터 프린트)
print(recommend)
```

```python
print('앉아서 가고 싶나? {0}:{1}에 열차를 타라.' .format(recommend[1:3], recommend[3:5]))
```
